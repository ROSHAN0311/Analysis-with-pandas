```python
df=df[['coaster_name', 'Location', 'Status', 
      'Manufacturer',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean','speed_mph', 'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df.rename(columns={'opening_date_clean':'Opening_Date_Clean',
                  'coaster_name':'Coaster_Name',
                   'latitude':'Latitude',
                   'longitude':'Longitude',
                   'year_introduced':'Year_Introduced',
                   'speed_mph':'Speed_Mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions_Clean'
                  },inplace=True)
```

```python
df =df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date_Clean'])] 
```

```python
df=df.reset_index(drop=True).copy()
```

```python
df.head(5)
```

# Graphical representation

```python
df['Year_Introduced'].value_counts()\
.head(10)\
.plot(kind='bar',color='g',title='Top 10 Coaster Introduced Years',xlabel='Years',ylabel='No. of Coaster Introduced')
plt.show()
```

```python
a=df["Speed_Mph"].plot(kind='hist',alpha=1,bins=20,title='Coaster Speed(Mph)')
a.set_xlabel("Speed(Mph)")
plt.show()
```

```python
a=df["Speed_Mph"].plot(kind='kde',alpha=1,title='Coaster Speed(Mph)')
a.set_xlabel("Speed(Mph)")
plt.show()
```

```python
df.plot(kind="scatter",
       x='Speed_Mph',
       y='Height_ft',
       title="Speed vs Height")
plt.show()
```

```python
sns.scatterplot(
      data=df,
        x='Speed_Mph',
        y='Height_ft',
        hue='Year_Introduced'
)
plt.show()
```

```python
sns.pairplot(df,vars=['Year_Introduced','Speed_Mph','Height_ft','Inversions_Clean','Gforce_clean'],
            hue='Type_Main'
            )
plt.show()
```

```python
df_corr=df[['Year_Introduced','Speed_Mph','Height_ft','Inversions_Clean','Gforce_clean']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
df.query('Location != "Other"')\
.groupby('Location')['Speed_Mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',figsize=(12,5),xlabel='No. of Roller Coaster')
plt.show()
```
